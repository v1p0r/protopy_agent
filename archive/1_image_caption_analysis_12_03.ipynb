```python
import json
import subprocess

from pathlib import Path
```

```python
input = Path("/Users/v1p0r/Documents/tmp/paperbench_data/data/adaptive-pruning")
content_list_path = next(input.glob("*/auto/*content_list.json"))
print(content_list_path)
content_list = json.loads(content_list_path.read_text())

print(content_list[0])
```

```python
image_pairs = {}
index = 0
for i in content_list:
    if i['type'] == 'image':

        image_pairs[index] = {
            "image": str(Path(next(input.glob("*/auto/"))).joinpath(i['img_path'])),
            "caption": " ".join(i['image_caption']).strip()
        }
        index += 1

print(image_pairs)


```

```python
for pdf in input.rglob("*.pdf"):
    out_dir = pdf.parent
    subprocess.run(
        ["mineru", "-p", str(pdf), "-o", str(out_dir)],
        check=True,
    )
```
