```python
def combine_pngs_to_base64(src_dir):
    """
    Reads all .png files (recursively) under src_dir,
    converts them to base64, and concatenates them into one long string.
    """
    all_b64 = ""
    for root, _, files in os.walk(src_dir):
        for file in sorted(files):
            if file.lower().endswith(".png"):
                path = os.path.join(root, file)
                with open(path, "rb") as f:
                    encoded = base64.b64encode(f.read()).decode("utf-8")
                    all_b64 += encoded
    return all_b64

src_dir = "/Users/v1p0r/Documents/DDL_Fall25/term_project/protopy_agent/image_test"
encoded_images = combine_pngs_to_base64(src_dir)
total_tokens = encoding.encode(len(encoded_images))

def image_to_base64(image_path: str) -> str:
    """
    Convert a single image file to a Base64 string.
    
    Args:
        image_path (str): Path to the image file (.png, .jpg, etc.)
    
    Returns:
        str: Base64-encoded text of the image
    """
    with open(image_path, "rb") as f:
        encoded_bytes = base64.b64encode(f.read())
    return encoded_bytes.decode("utf-8")

src_images = "/Users/v1p0r/Documents/DDL_Fall25/term_project/protopy_agent/pdf_test/besiegefield_pipeline_compressed-000001.png"
encoded_image = image_to_base64(src_images)
print(len(encoding.encode(encoded_image)))

```

```python
import os
import base64
import tiktoken

def count_tokens_for_file(encoding, file_path):
    """Count tokens for a single image (Base64 encoded)."""
    with open(file_path, "rb") as f:
        b64 = base64.b64encode(f.read()).decode("utf-8")
    return len(encoding.encode(b64))

def count_total_tokens_in_folder(folder_path, encoding_name="cl100k_base"):
    """
    Loop through all image files in a folder and sum token counts.
    Supports recursive search.
    """
    encoding = tiktoken.get_encoding(encoding_name)
    total_tokens = 0
    per_file = {}

    for root, _, files in os.walk(folder_path):
        for file in sorted(files):
            if file.lower().endswith((".png", ".jpg", ".jpeg", ".webp")):
                path = os.path.join(root, file)
                tokens = count_tokens_for_file(encoding, path)
                per_file[path] = tokens
                total_tokens += tokens
    
    print("\n=== Token counts per file ===")
    for f, t in per_file.items():
        print(f"{f}: {t} tokens")
    print(f"\nTotal tokens across folder: {total_tokens}")
    return total_tokens, per_file

print("Interested figure token size: 108718")
# Example usage:
src_dir = "/Users/v1p0r/Documents/DDL_Fall25/term_project/protopy_agent/image_test"
total_tokens, details = count_total_tokens_in_folder(src_dir)



print("Decrease the token number by :", 1-(108718/39602252))
```
